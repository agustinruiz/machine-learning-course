### Matplotlib example workflow

```python
# 0. import matplotlib and getting ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # figsize establece el ancho y alto del plot. (width,height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show (you save the hold figure)
fig.savefig("images/sample-plot.png")
```

### Bar Plot

```python
# Use some data to make a bar plot
# Make a plot from dictionary
nut_butter_prices = {"Almond butter":10,
                     "Peanut butter":8,
                     "Cashew butter":12}
fix, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store",
      ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

### Histogram

```python
# Make some data for histograms and plot it
x = np.random.randn(1000) # Esto me dara 1000 numeros aleatorios dentro de la distribucion normal. (con una distrubucion normal)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subplots. Several plots in the same image

```python
# Subplot option 1 (the recomended)
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))

# Plot to each different axis
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Plotting from Pandas DataFrames

```python
import pandas as pd

# Make a dataFrame
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
# This is a way to create a date Serie where a random number is asigned to each date
ts = pd.Series(np.random.randn(1000),
              index=pd.date_range(1/1/2021,periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
# Data manipulation. Format de price column in integer format.
# In the text i have to replace de dolar simbol, the coma and the dot for an empty string
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
# Now we have to remove the las two zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
# Now we are going to add an aditional column for the sale dates
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=len(car_sales))
car_sales
```

```python
# now we are going to create a new column with the total sales
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# because price is in string format the cumsum concatenate the strings
type(car_sales["Price"][0])
```

```python
# we have to conver the price column into an integer
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# Let's plt the total sales
car_sales.plot(x="Sale Date",y="Total Sales");
```

```python
# Reassign Price column to int
car_sales["Price"] = car_sales["Price"].astype(int)
# Plot scatter plot with price collumn as numeric
car_sales.plot(x="Odometer (KM)",y="Price", kind="scatter");
```

### Wich method you slould use? (pyplot vs matplotlib oo method)

* When plotting something quickly, ok to use the pyplot method
* When plotting something more advanced, use the oo method

```python
heart_disease.head()
```

```python
# Si quiero solo los pacientes mayores de 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method. Plotting directli from the pandas DataFrame
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO Method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
# we see that just addin the ax the x label apear
# now with ax we could modify any part of the plot
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

ax.set_xlim([45,100]) # modify the limits of the x axis
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target") # * means unpack all the values. in this case: "Grab the values from scatter and unpack the legend_elements"

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of thol, age and thalach
# voy a crear una figura que tenga dos pares de ejes
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10)
                              ,sharex=True # Con esta linea le digo al grafico que ambas ax van a compartir el eje x=> puedo remover el xlabel del ax0
                              )

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
#Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
#       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add Meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use("seaborn-whitegrid")

# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") # This changes the color scheme

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target") # * means unpack all the values. in this case: "Grab the values from scatter and unpack the legend_elements"

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
# customizing the y and x axis limitations


# Subplot of thol, age and thalach
# voy a crear una figura que tenga dos pares de ejes
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10)
                              ,sharex=True # Con esta linea le digo al grafico que ambas ax van a compartir el eje x=> puedo remover el xlabel del ax0
                              )

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")
#Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
#       xlabel="Age",
       ylabel="Cholesterol")
# Change the x axis limits
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                      cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")
# Change ax1 axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add Meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
